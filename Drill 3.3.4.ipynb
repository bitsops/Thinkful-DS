```python
import numpy as np
import pandas as pd
import scy
import matplotlib.pyplot as plt
%matplotlib inline
```

# binomial

```python
distribution_binomial = np.random.binomial(1,.5,100)
```

```python
distribution_binomial
plt.hist(distribution_binomial)
plt.axvline(distribution_binomial.mean(), color='b', linestyle='solid')
plt.axvline(distribution_binomial.mean() + distribution_binomial.std(), color='b', linestyle='dashed')
plt.axvline(distribution_binomial.mean() - distribution_binomial.std(), color='b', linestyle='dashed')
plt.show
```

# Normal Distribution Section

```python
rand1 = np.random.normal(5,.5,100000)
rand2 = np.random.normal(10,1,100000)
rand3 = rand1+rand2



plt.hist(rand3)
plt.axvline(rand3.mean(), color='b', linestyle='solid')
plt.axvline(rand3.mean() + rand3.std(), color='b', linestyle='dashed')
plt.axvline(rand3.mean() - rand3.std(), color='b', linestyle='dashed')
plt.show

data = pd.DataFrame()
data['rand1'] = rand1
data['rand2'] = rand2
data['rand3'] = rand3

data.describe()
```
